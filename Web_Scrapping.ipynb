```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
flipkart = requests.get("https://www.flipkart.com/search?q=realme+Mobile")
```

```python
flipkart.content
```

```python
soup = BeautifulSoup(flipkart.content, "html.parser")
```

```python
soup
```

```python
soup.title
```

```python
soup.p
```

```python
soup.div
```

```python
soup.div.text
```

```python
mobile = soup.find("div" , class_="_13oc-S")
mobiles = soup.find_all("div" , class_="_13oc-S")
```

```python
len(mobile)
```

```python
name = mobile.find("div" , class_="_4rR01T")
name.text
```

```python
price = mobile.find("div" , class_= "_30jeq3 _1_WHN1")
price.text
```

```python
ram = mobile.find_all("li")[0]
ram.text
```

```python
ram.text.split("|")[0]  
```

```python
name_arr = []
price_arr = []
ram_arr = []
#url_arr = []

for mobile in mobiles:
  name = mobile.find("div" , class_="_4rR01T").text
  price = mobile.find("div" , class_= "_30jeq3 _1_WHN1").text
  ram = mobile.find_all("li")[0].text.split(" | ")[0]
  #url = mobile.find("a" , class_="_1fQZEK")
  name_arr.append(name);
  price_arr.append(price);
  ram_arr.append(ram);
  #url_arr.append(url)
```

```python
name_arr
```

```python
dist = {"Model" : name_arr , "Price" : price_arr , "RAM" : ram_arr }
```

```python
import pandas as pd
df = pd.DataFrame(dist)
```
